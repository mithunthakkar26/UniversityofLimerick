```
filePath1 = "/content/positive-words.txt"
filePath2 = "/content/negative-words.txt"

with open(filePath1) as f:
  Positive_Words = f.read().splitlines()[35:]

with open(filePath2, errors='replace') as f:
  Negative_Words = f.read().splitlines()[35:]
```

```
def sentimentAnalyzer(string):
  print("string = ", string)
  Count_Positive = 0
  Count_Negative = 0
  String_List = string.split(" ")
  for i in String_List:
    if i in Positive_Words:
      Count_Positive+=1
    elif i in Negative_Words:
      Count_Negative+=1

    
  
  if Count_Positive>Count_Negative:
    print("Positive Sentiment \U0001F642")
    print("Confidence:", Count_Positive/len(String_List))
  elif Count_Positive<Count_Negative:
    print("Negative Sentiment \U0001F641")
    print("Confidence:", Count_Negative/len(String_List))
  else:
    print("Neutral Sentiment \U0001F610")
    print("Confidence:", Count_Positive/len(String_List))

      


sentimentAnalyzer("NLP is cool")
sentimentAnalyzer("NLP is cool and useful")
sentimentAnalyzer("NLP is hard")
sentimentAnalyzer("NLP is hard and useless")
```

```
trainingSet = [("London is the capital of GB", "GB"), ("Oxford is a city in GB", "GB"), ("Dublin is the capital of Ireland", "IE"), ("Limerick is a city in Ireland","IE")]
testSet = [("University of Limerick", "?"), ("University College Dublin", "?"), ("Imperial College London", "?"), ("Ireland & GB", "?")]

def NaiveBayesClassifier(trainingSet, testSet):
  megaDocGB = ""
  megaDocIE = ""
  countGB = 0
  countIE = 0
  GB_List = []
  IE_List = []
  vocab = []
  for i in trainingSet:
    if i[1] == "GB":
      megaDocGB = megaDocGB+" "+i[0]
      countGB +=1
      for j in i[0].split(" "):
        GB_List.append(j)
        vocab.append(j)
    else: 
      megaDocIE = megaDocIE+" "+i[0]
      countIE +=1
      for j in i[0].split(" "):
        IE_List.append(j)
        vocab.append(j)
  print("megaDocGB=", megaDocGB)
  print("megaDocIE=",megaDocIE)

  probGB = countGB/(countGB+countIE)
  probIE = countIE/(countGB+countIE)

  print("probGB = ", probGB)
  print("probIE = ", probIE)

  vocab = set(vocab)

  GB_BOW = {}
  IE_BOW = {}
  for i in set(GB_List):
    GB_BOW[i] = GB_List.count(i)
  print("GB_BOW = ", GB_BOW)

  for i in set(IE_List):
    IE_BOW[i] = IE_List.count(i)
  print("IE_BOW=", IE_BOW)

  

  print("---------------------------------------------")
  for i in testSet:
    print("Test Document1:", i)
    docProbGB = probGB
    docProbIE = probIE
    for j in i[0].split(" "):
      if j in GB_List:
        condProb = (GB_BOW[j]+1)/(len(GB_List)+len(vocab)+1)
        print("word = ", j, ", Word in GB list, ", "wordConditionalProbabilityGB = ", condProb)
        docProbGB = docProbGB*condProb
      elif j not in GB_List and j in IE_List:
        condProb =(1/(len(GB_List)+len(vocab)+1))
        print("word = ", j, ", Word not in GB List, ", "wordConditionalProbabilityGB = ", condProb)
        docProbGB = docProbGB*condProb
      else: 
        condProb =(1/(len(GB_List)+len(vocab)+1))
        print("word = ", j,", Unknown Word, ","wordConditionalProbabilityGB = ", condProb)
        docProbGB = docProbGB*condProb

      if j in IE_List:
        condProb =(IE_BOW[j]+1)/(len(IE_List)+len(vocab)+1)
        print("word = ", j, ", Word in IE List, ", "wordConditionalProbabilityIE = ", condProb)
        docProbIE = docProbIE*condProb
      elif j not in IE_List and j in GB_List:
        condProb =(1/(len(IE_List)+len(vocab)+1))
        print("word = ", j, ", Word not in IE list, ", "wordConditionalProbabilityIE = ", condProb)
        docProbIE = docProbIE*condProb
      else:
        condProb =(1/(len(IE_List)+len(vocab)+1))
        print("word = ", j , ", Unknown Word, " , "wordConditionalProbabilityIE = ", condProb)
        docProbIE = docProbIE*condProb
      print("\n")
    
    print("docProbGB: ", format(docProbGB, '0.8f'))
    print("docProbIE: ", format(docProbIE, '0.8f'))

    if round(docProbGB,8)>round(docProbIE,8):
      print("Inferred Class: GB")
    elif round(docProbGB,8)<round(docProbIE,8):
      print("Inferred Class: IE")
    elif round(docProbGB,8)==round(docProbIE,8):
      print("Inferred Class: Both")

    print("---------------------------------------------")

    



NaiveBayesClassifier(trainingSet, testSet)
```
