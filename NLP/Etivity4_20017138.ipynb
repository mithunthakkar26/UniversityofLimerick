```
def nonWordSpellingCorrection(nonWord):
  candidates = []
  for i in unigrams_df.unigram:
    try:
      if jellyfish.damerau_levenshtein_distance(nonWord, i)<2 and i is not None:
        candidates.append([i, unigrams_df.loc[unigrams_df.unigram==i]["count"].values[0], '{0:.8f}'.format(unigrams_df.loc[unigrams_df.unigram==i]["count"].values[0]/unigrams_df["count"].sum())])
    except TypeError:
      pass
  df = pd.DataFrame(candidates, columns =['word', 'frequency', 'P(word)']) 
  return df

df = nonWordSpellingCorrection("acress")
df
  
```
