```
import pandas as pd
 
!wget https://norvig.com/ngrams/count_1w.txt #unigram corpus
!wget https://norvig.com/ngrams/count_2w.txt #bigram corpus
 
filePath1 = "/content/count_1w.txt"
filePath2 = "/content/count_2w.txt"
 
unigrams_df = pd.read_csv(filePath1,sep='\t',header=None, names=['unigram','count'])
bigrams_df = pd.read_csv(filePath2,sep='\t',header=None, names=['bigram','count'])
 
print(f'Number of unigrams: {unigrams_df.size} Number of Bigrams: {bigrams_df.size}')
 
display(unigrams_df.head(100))
display(bigrams_df.head(100))
```

```
bigrams_prob = 1
def probability(sentence, bigrams_prob):
  sentence2 = sentence.split()
  if len(sentence2)>1:
    i = 0
    print(f"Unigram count for {sentence2[i]}: ", unigrams_df.loc[unigrams_df.unigram==sentence2[i]]["count"].values[0])
    print(f"Bigram count for {sentence2[i]} {sentence2[i+1]}: ", bigrams_df.loc[bigrams_df.bigram==sentence2[i]+" "+sentence2[i+1]]["count"].values[0])  
    prob = bigrams_df.loc[bigrams_df.bigram==sentence2[i]+" "+sentence2[i+1]]["count"].values[0]/unigrams_df.loc[unigrams_df.unigram==sentence2[i]]["count"].values[0]
    print(f"Bigram probability for {sentence2[i]} {sentence2[i+1]}: ", prob)  
    print("----------------------------------")
    sentence = " ".join(sentence2[i+1:])
    bigrams_prob = bigrams_prob*prob
    probability(sentence, bigrams_prob)
    print("----------------------------------")
  else: 
    sentenceProbability = format(float(bigrams_prob),'.25f')
    print(f"Sentence probability: ", sentenceProbability)
    print("----------------------------------")
  sentenceProbability = bigrams_prob
  return sentenceProbability
 
probability("i love you", bigrams_prob)>probability('i hate you', bigrams_prob)
```
