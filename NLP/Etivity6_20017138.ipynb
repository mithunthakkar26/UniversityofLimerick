```
from textblob import TextBlob
```

```
def sentimentAnalyzer(string):
  print("String = ", string)
  tb = TextBlob(string)
  print(tb.sentiment)
  if tb.sentiment.polarity<0:
    print("Negative Sentiment \U0001F641")
  elif tb.sentiment.polarity>0:
    print("Positive Sentiment \U0001F642")
  else: 
    print("Neutral Sentiment \U0001F610")

  print(f"Subjectivity: {round(tb.sentiment.subjectivity,2)}")

sentimentAnalyzer("NLP is cool")
sentimentAnalyzer("NLP is cool and useful")
sentimentAnalyzer("NLP is hard")
sentimentAnalyzer("NLP is hard and useless")
sentimentAnalyzer("NLP is there")
```

```
from textblob.classifiers import NaiveBayesClassifier

trainingSet = [("London is the capital of GB", "GB"), ("Oxford is a city in GB", "GB"), ("Dublin is the capital of Ireland", "IE"), ("Limerick is a city in Ireland","IE")]
testSet = [("University of Limerick", "IE"), ("University College Dublin", "IE"), ("Imperial College London", "GB"),("University of Oxford", "GB"), ("Ireland & GB", "GB")]

def naiveBayesClassifier(trainingSet, testSet):
  cl = NaiveBayesClassifier(trainingSet)
  correctClassification = 0
  for i in testSet:
    print(f"document:{i}      predicted class: {cl.classify(i[0])}")
    if i[1]== cl.classify(i[0]):
      correctClassification += 1
  print(f"Accuracy: {correctClassification/len(testSet)}")



naiveBayesClassifier(trainingSet, testSet)
```

```
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
```

```
display(df.head())
print("\n")
print(f"Number of sample documents in the dataset:{len(df)}")
print("\n")
print(df['category'].value_counts())
```

```
df = df.set_index("text")
```

```
TrainSize = round(0.8*len(df))
TestSize = len(df) - TrainSize
TrainingSet = df.iloc[0:TrainSize,:]
TestSet = df.iloc[0:TestSize,:]
```

```
TrainingSet = TrainingSet.to_records()
TestSet = TestSet.to_records()
```

```
cl = NaiveBayesClassifier(TrainingSet)
cl.accuracy(TestSet)
```
