```
import time
import numpy as np
from tabulate import tabulate
```

```
def edDistanceRecursive(x,y):
  if len(x)==0: return len(y)
  if len(y)==0: return len(x)
  delt = 2 if x[-1]!=y[-1] else 0
  diag = edDistanceRecursive(x[:-1], y[:-1])+delt
  vert = edDistanceRecursive(x[:-1], y)+1
  horz = edDistanceRecursive(x, y[:-1])+1
  return min(diag,vert,horz)
```

```
start = time.time()
str1 = "INTENTION"
str2 = "EXECUTION"
print(f'Edit Distance between "{str1}" and "{str2}": {edDistanceRecursive(str1, str2)}')
end = time.time()
print("----",end - start, "seconds---")
```

```
def editDistDP(str1, str2, m, n):
  # Create a table to store results of subproblems
  dp = [[0 for x in range(n + 2)] for x in range(m + 2)]
  dp_BT = [["" for x in range(n + 2)] for x in range(m + 2)]
  # Fill dp[][] in bottom up manner
  for i in range(2,m+2):
    dp[i][0] = str1[i-2]

  for j in range(2,n+2):
    dp[0][j] = str2[j-2]

  for i in range(1,m + 2):
    for j in range(1,n + 2):
      # If first string is empty, only option is to
      # insert all characters of second string
      if i == 1:
        dp[i][j] = j-1    # Min. operations = j
      # If second string is empty, only option is to
      # remove all characters of second string
      elif j == 1:
        dp[i][j] = i-1    # Min. operations = i
      # If last characters are same, ignore last char
      # and recur for remaining string
      elif str1[i-2].lower() == str2[j-2].lower():
          dp[i][j] = dp[i-1][j-1]
          dp_BT[i][j]+="↖"
      # If last character are different, consider all
      # possibilities and find minimum
      else:
        dp[i][j] = min(int(dp[i-1][j])+1,        # Insert
                            int(dp[i][j-1])+1,        # Remove
                            int(dp[i-1][j-1])+2)    # Replace
        arrows = {"←":int(dp[i][j-1]), "↑":int(dp[i-1][j]), "↖":int(dp[i-1][j-1]+1)}
        for k,l in arrows.items():
          if l==min(int(dp[i-1][j]),int(dp[i][j-1]),int(dp[i-1][j-1]+1)):
            dp_BT[i][j]+=k
        
  
  for i in range(1,m + 2):
    for j in range(1,n + 2):
        if i == 1:
          dp[i][j]=str(dp[i][j])+"←"
        elif j == 1:
          dp[i][j]= str(dp[i][j])+"↑"


  return dp, dp_BT

 # Driver program
str1 = "Intention"
str2 = "Execution"
start = time.time()
dp,dp_BT = editDistDP(str1, str2, len(str1), len(str2))
from tabulate import tabulate
print(tabulate(np.char.add(dp_BT,dp)))
end = time.time()
print("----",end - start, "seconds---")
```
