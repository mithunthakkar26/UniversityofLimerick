```python
class Determinant_Inverse:
    def __init__(self, Matrix):
        self.Matrix = Matrix
        
    def Determinant_2by2(self):
        try:
            return self.Matrix[0][0]*self.Matrix[1][1]-self.Matrix[0][1]*self.Matrix[1][0]
        except:
            return "Matrix should be 2by2"
    
    def Compute_Inverse_Matrix(self):
        det = self.Determinant_2by2()
        try: 
            New_Matrix = [self.Matrix[1][1], -self.Matrix[0][1]], [-self.Matrix[1][0],self.Matrix[0][0]]
            Inverse_Matrix = [[round(1/det*New_Matrix[0][0],2), round(1/det*New_Matrix[0][1],2)],[round(1/det*New_Matrix[1][0],2), round(1/det*New_Matrix[1][1],2)]]
            return Inverse_Matrix
        except ZeroDivisionError: 
            print("Matrix is not invertible")

A = [[3,8],[4,2]]
di = Determinant_Inverse(A)
det = di.Determinant_2by2()
A_Inv = di.Compute_Inverse_Matrix()
print(det,A_Inv)
B = [[3,8],[3,8]]
di = Determinant_Inverse(B)
det = di.Determinant_2by2()
B_Inv = di.Compute_Inverse_Matrix()
print(det,B_Inv)
```

```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
%matplotlib inline 
```

```python
A = [[4,3,3],[2,6,1],[2,3,6]]
def Plot_3DLinePlot(Matrix, Axis_Scale):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    Column_Matrix = [[Matrix[0][0],Matrix[1][0], Matrix[2][0]],[Matrix[0][1],Matrix[1][1], Matrix[2][1]],[Matrix[0][2],Matrix[1][2], Matrix[2][2]]]
    ax.quiver(0,0,0,Column_Matrix[0],Column_Matrix[1],Column_Matrix[2])
    ax.set_xlim(0,Axis_Scale)
    ax.set_ylim(0,Axis_Scale)
    ax.set_zlim(0,Axis_Scale)
    plt.show()
Plot_3DLinePlot(A,10)
```

```python
A = [[4,3,3],[2,6,1],[2,3,6]]
B = [[3,6,7],[2,3,5],[8,5,7]]
C = [[ 0 for i in range(3)] for j in range(3)]
for i in range(3):
   for j in range(3):
       for k in range(3):
           C[i][j] += A[i][k] * B[k][j]
Plot_3DLinePlot(C, 60)
```
