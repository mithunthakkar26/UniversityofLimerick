```python
import random 
import numpy as np
import matplotlib 
import math 
import sklearn
import pandas as pd
```

```python
# Check if a coin is fair coin
p = 0.5
n = 100
simulations = 10
s_heads=[]
s_tails = []
#below loop is feed experiment results into s_heads and s_tails lists
for _ in range(simulations):
    heads = 0
    tails = 0
    #below loop is for finding out number of heads and tails obtained after 100 coin tosses
    for i in range(n):
        coin_toss = np.random.binomial(1,p)
        if coin_toss == 1:
            heads+=1
        else:
            tails+=1
    s_heads.append(heads)
    s_tails.append(tails)

#below mean and standard deviations are calculated based on mean and standard deviation formula for binomial distributions
mean = n*p
std_dev = np.sqrt(mean*(1-p))
print(f"Expected number of heads or tails in a fair coin: {mean}")
print(f"Expected Standard deviation: {std_dev}")
print(f"Mean Results: heads: {np.mean(s_heads)} and tails: {np.mean(s_tails)}")
print(f"Standard Deviation Results: heads: {np.std(s_heads)} and tails: {np.std(s_tails)}")
```

```python
plt.hist(s_heads, bins = simulations)
plt.show()
```

```python
# Checking if weights of people in Limerick City have weights around 150 lbs
people = 50
simulation_weights = []
simulations = 200
mean_weight = 150
std_weight = 3
#below loop is feed experiment results into simulation_weights list
for _ in range(simulations):
    total_weight = 0
    #below loop is for generating normally distributed random weights and totalling the weights obtained
    for i in range(1,people):
        person_weight = random.normalvariate(mean_weight,std_weight)
        total_weight+=person_weight
    average_weight = total_weight/people
    simulation_weights.append(average_weight)

print(f"Expected Value of the experiment: {random.choice(simulation_weights)}")
print(f"Standard Deviation: {np.std(simulation_weights)}")
```

```python
plt.hist(simulation_weights, bins = simulations)
plt.show()
```
